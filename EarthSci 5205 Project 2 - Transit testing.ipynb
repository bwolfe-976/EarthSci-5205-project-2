```python
import matplotlib.pyplot as plt
import numpy as np
import astropy.io.ascii
plt.ion()
%matplotlib inline
```

```python
tr_data = astropy.io.ascii.read("UID_0108859_PLC_003.tbl.txt", data_start=131, delimiter='|')
#data for HD209458 b transit data 
#data from Exoarchive Optical(Most) Rowe et al 2006
#data starts at line 130
```

```python
flx = 2.512**(-tr_data["Relative_Magnitude"])
flx = flx / np.median(flx)
plt.errorbar(tr_data["JD"], flx, tr_data["Relative_Magnitude_Uncertainty"], ms=5, \
             marker="o", ls="none", label="Data", color="black", alpha=0.3)
```

```python
t_tr = tr_data["JD"].data
per = 3.5247485
ind = np.argsort(t_tr % per)
plt.errorbar(tr_data["JD"] %per, flx, tr_data["Relative_Magnitude_Uncertainty"], ms=5, \
             marker="o", ls="none", label="Data", color="black", alpha=0.3)
plt.xlabel("Time [day]")
plt.ylabel("Transit Depth/Relative Mag")

```
